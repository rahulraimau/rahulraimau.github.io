## Bernoulli Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.7
data = bernoulli.rvs(p, size=1000)
df = pd.DataFrame(data, columns=["Bernoulli"])
df["Bernoulli"].value_counts().plot(kind="bar", title="Bernoulli Distribution")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
df.head()
```

## Binomial Distribution

```python
from scipy.stats import binom

n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)
df = pd.DataFrame(data, columns=["Binomial"])
df["Binomial"].plot.hist(bins=n+1, rwidth=0.9, title="Binomial Distribution")
plt.xlabel("Number of Successes")
plt.grid(True)
plt.show()
df.head()
```

## Poisson Distribution

```python
from scipy.stats import poisson

mu = 4
data = poisson.rvs(mu, size=1000)
df = pd.DataFrame(data, columns=["Poisson"])
df["Poisson"].plot.hist(bins=30, rwidth=0.9, title="Poisson Distribution")
plt.xlabel("Event Count")
plt.grid(True)
plt.show()
df.head()
```

## Uniform Distribution

```python
from scipy.stats import uniform

a, b = 0, 10
data = uniform.rvs(loc=a, scale=b-a, size=1000)
df = pd.DataFrame(data, columns=["Uniform"])
df["Uniform"].plot.hist(bins=30, title="Uniform Distribution")
plt.xlabel("Value")
plt.grid(True)
plt.show()
df.head()
```

## Normal Distribution

```python
from scipy.stats import norm

mu, sigma = 0, 1
data = norm.rvs(loc=mu, scale=sigma, size=1000)
df = pd.DataFrame(data, columns=["Normal"])
df["Normal"].plot.kde(title="Normal Distribution")
plt.grid(True)
plt.show()
df.head()
```

## Exponential Distribution

```python
from scipy.stats import expon

scale = 1
data = expon.rvs(scale=scale, size=1000)
df = pd.DataFrame(data, columns=["Exponential"])
df["Exponential"].plot.hist(bins=30, title="Exponential Distribution")
plt.grid(True)
plt.show()
df.head()
```

## Student's t-Distribution

```python
from scipy.stats import t

df_t = 10
data = t.rvs(df_t, size=1000)
df = pd.DataFrame(data, columns=["T"])
df["T"].plot.kde(title="Student's t-Distribution")
plt.grid(True)
plt.show()
df.head()
```

## Chi-Square Distribution

```python
from scipy.stats import chi2

df_chi = 5
data = chi2.rvs(df_chi, size=1000)
df = pd.DataFrame(data, columns=["Chi-Square"])
df["Chi-Square"].plot.hist(bins=30, title="Chi-Square Distribution")
plt.grid(True)
plt.show()
df.head()
```

## Gamma Distribution

```python
from scipy.stats import gamma

shape, scale = 2, 1
data = gamma.rvs(a=shape, scale=scale, size=1000)
df = pd.DataFrame(data, columns=["Gamma"])
df["Gamma"].plot.hist(bins=30, title="Gamma Distribution")
plt.grid(True)
plt.show()
df.head()
```

## Beta Distribution

```python
from scipy.stats import beta

a, b = 2, 5
x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
plt.plot(x, y, label=f"Beta({a},{b})")
plt.title("Beta Distribution PDF")
plt.legend()
plt.grid(True)
plt.show()
```

# Naïve Bayes Classifiers – Gaussian, Multinomial, Bernoulli

Naïve Bayes classifiers are a family of simple probabilistic models based on Bayes' Theorem. 
They assume independence between features and work well for high-dimensional datasets like text classification.

Bayes' Theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

 Gaussian Naïve Bayes – For continuous features (e.g., Iris dataset)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

# Load data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)

# GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

2️⃣ Multinomial Naïve Bayes – For count-based text features

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample text
texts = ["I love data science", "Data science is great", "I hate bugs", "Bugs are annoying"]
labels = [1, 1, 0, 0]  # 1 = Positive, 0 = Negative

# Vectorize text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Train model
mnb = MultinomialNB()
mnb.fit(X, labels)

# Predictions
preds = mnb.predict(X)
print("Predictions:", preds)
```

Bernoulli Naïve Bayes – For binary features (e.g., word present or not)

```python
from sklearn.naive_bayes import BernoulliNB

# Convert to binary (presence/absence)
X_binary = (X > 0).astype(int)

# Train model
bnb = BernoulliNB()
bnb.fit(X_binary, labels)

# Predictions
print("Predictions:", bnb.predict(X_binary))
```

- Use **GaussianNB** for normally distributed features.
- Use **MultinomialNB** for discrete counts (e.g., word frequencies).
- Use **BernoulliNB** for binary features (e.g., 0/1 indicators).
- These models are **fast, interpretable**, and work well on high-dimensional text data.
