```python
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({i: s for i, s in enumerate(iris.target_names)})

# Show sample
print(df.head())
print(df['species'].value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=1)

# Model with Gini
tree_gini = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=1)
tree_gini.fit(X_train, y_train)

# Visualize
plt.figure(figsize=(12, 8))
plot_tree(tree_gini, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree (Gini)")
plt.show()
```
